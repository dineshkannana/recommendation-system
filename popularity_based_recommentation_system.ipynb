```python
import pandas as pd

# List of genres
genres = ['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
          'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
          'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

# Create an empty list to store rows
rows = []

# Loop through each genre and collect top 5 movie titles
for genre in genres:
    top5 = items[items[genre] == 1][['movie title']].head(5)
    for title in top5['movie title']:
        rows.append({'Genre': genre, 'Movie Title': title})

# Create DataFrame
top5_movies_df = pd.DataFrame(rows)

# Show the result
print(top5_movies_df)
```

```python
horizontal_movies_df = top5_movies_df.groupby('Genre')['Movie Title'].apply(list).to_frame()
```
